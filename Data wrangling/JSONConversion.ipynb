```python
keep_columns = ["Title", "Director", "Genre", "Plot", "imdbRating", "imdbVotes", "Poster"]

merged_df = pd.DataFrame(columns=keep_columns)

for infile_name in ["csv_generation/merged_movie_data_success.csv", "csv_generation/merged_movie_data_success2.csv", "csv_generation/merged_movie_data_success3.csv", "csv_generation/merged_movie_data_success4.csv", "csv_generation/merged_movie_data_success5.csv"]:
    infile = pd.read_csv(infile_name, usecols=keep_columns)
    merged_df = pd.concat([merged_df, infile])

merged_df.drop_duplicates(inplace=True)
merged_df.dropna(subset=["Plot"], inplace=True)

merged_df.to_csv("final_5k_movies_data.csv", index=False)
merged_df.to_json("merged_movie_data.json", orient="records")
```
